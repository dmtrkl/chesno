### Можливі запити 

#### Місцеві ради та обрані депутати, що не входили до початкового списку.

```python
from IPython.display import display 
sub = set(winners['ПІБ']) - set(candidates['ПІБ'])
display(winners[winners['ПІБ'].isin(sub)].head())
print("Amount of winners who was not in the candidate list: ", len(sub))
print("Additional councils:")
for r in (set(winners['Рада'].unique()) - set(candidates['Рада'].unique())):
    print('\t', r)
```

#### Кандидати, що подавались на декілька позицій.

```python
# Candidates applying to several positions
candidates[candidates.duplicated(keep=False, subset=['ПІБ','Дата народження'])].sort_values(by='ПІБ').head()
```

#### Кандидати з однаковими ПІБ, що подавались до однієї ради.

```python
# Candidates with duplicated names applying for the same council
candidates[candidates.duplicated(keep=False, subset=['ПІБ','Рада'])].sort_values(by='ПІБ').head()
```

#### Кандидати, які представлені більш ніж однією партією.

```python
# Candidates nominated by more than one party
candidates[candidates.duplicated(keep=False, subset=['ПІБ','Дата народження'])]\
    .groupby('ПІБ')\
    .filter(lambda g: (g['Партія'].nunique() > 1))
```

#### Кільскість депутатів, що обрано більш ніж на одну посаду.

```python
print(
    "Amount of candidates elected for several positions : ",
    len(winners[winners.duplicated(keep=False, subset=['ПІБ','Дата народження'])].sort_values(by='ПІБ'))
)
```
